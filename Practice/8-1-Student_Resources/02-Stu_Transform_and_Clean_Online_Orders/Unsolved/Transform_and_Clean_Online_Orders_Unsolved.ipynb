```python
import pandas as pd
```

```python
# Read in the data
pd.options.display.max_colwidth = 200
orders_df = pd.read_csv("Resources/online_orders.csv")
orders_df
```

```python
# Create a list for the column names.
columns = orders_df.columns.to_list()
columns
```

```python
# Split the column on the comma and place in a list.
for column in columns:
    column_names = column.split(",")
print(column_names)
```

```python
# Create a new DataFrame and split the data from one column on the comma into six columns.
online_orders_df = pd.DataFrame()
online_orders_df[['orderId', 'productName', 'category', 'sellingPrice', 'shipping', 'totalPrice']] = orders_df['orderId,productName,category,sellingPrice,shipping,totalPrice'].str.split(',', n=6, expand=True)
online_orders_df
```

```python
# Convert sellingPrice and totalPrice to a float datatype.
online_orders_df[['sellingPrice','totalPrice']] = online_orders_df[['sellingPrice','totalPrice']].astype('float')
online_orders_df
```

```python
# Check the data types.
online_orders_df.info()
```

```python
# Calculate the shipping price of each product and add it to a new column.
online_orders_df['shippingPrice'] = online_orders_df['totalPrice'] - online_orders_df['sellingPrice']
online_orders_df
```

```python
# Reorder the columns ['orderId', 'productName', 'category', 'sellingPrice','shippingPrice', 'totalPrice', 'shipping']
online_orders_df = online_orders_df[['orderId', 'productName', 'category', 'sellingPrice','shippingPrice', 'totalPrice', 'shipping']]
online_orders_df
```
